```python
!pip install transformers gradio

from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import gradio as gr

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
model = AutoModelForCausalLM.from_pretrained("ibm-granite/granite-3.3-2b-instruct")
model.eval()
if torch.cuda.is_available():
    model.to('cuda')

def chatbot(user_input, chat_history=[]):
    # Add user input to message list
    chat_history.append({"role": "user", "content": user_input})

    # Prepare input for the model
    inputs = tokenizer.apply_chat_template(
        chat_history,
        add_generation_prompt=True,
        tokenize=True,
        return_dict=True,
        return_tensors="pt",
    )

    if torch.cuda.is_available():
        inputs = {k: v.to('cuda') for k, v in inputs.items()}

    # Generate output tokens
    outputs = model.generate(**inputs, max_new_tokens=100)

    # Extract generated response (skip input tokens)
    generated = tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:])

    # Append assistant's response to chat history
    chat_history.append({"role": "assistant", "content": generated.strip()})

    # Return latest assistant reply and updated history for next input
    return generated.strip(), chat_history

# Gradio interface
with gr.Blocks() as demo:
    chat_history = gr.State([])
    chatbot_ui = gr.Chatbot()
    user_input = gr.Textbox(placeholder="Type your message here...")
    submit_btn = gr.Button("Send")

    def user_submit(msg, history):
        response, history = chatbot(msg, history)
        # Update the chatbot display with user message and bot response
        chat_messages = []
        for m in history:
            if m["role"] == "user":
                chat_messages.append((m["content"], None))
            else:
                chat_messages[-1] = (chat_messages[-1][0], m["content"])
        return chat_messages, history, ""

    submit_btn.click(user_submit, inputs=[user_input, chat_history], outputs=[chatbot_ui, chat_history, user_input])
    user_input.submit(user_submit, inputs=[user_input, chat_history], outputs=[chatbot_ui, chat_history, user_input])

demo.launch()
```
